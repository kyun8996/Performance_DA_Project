```python
import os
import pandas as pd

# 상대 경로로 Excel 파일 경로 설정
file_path = os.path.join("Data", "퍼포먼스_마케팅_데이터.xlsx")

# 각 시트를 별도의 DataFrame으로 불러오기
df_performance = pd.read_excel(file_path, sheet_name="performance")
df_mmp = pd.read_excel(file_path, sheet_name="MMP")
```

```python
# index 컬럼에 중복값이 있는지 확인 (True/False)
is_duplicate = df_performance['index'].duplicated().any()

print("중복 여부:", is_duplicate)
```

```python
# 중복된 index 값만 추출
duplicated_values = df_performance[df_performance['index'].duplicated()]

print("중복된 index 값들:")
print(duplicated_values)
```

```python
# 컬럼명 일괄적으로 MySQL 테이블에 맞게 변환
df_performance = df_performance.rename(columns={
    "index": "index",
    "Channel": "channel",
    "Date": "date",
    "week": "week",
    "Campaign": "campaign",
    "Ad Group": "ad_group",
    "Ad Name": "ad_name",
    "Impressions": "impressions",
    "Clicks": "clicks",
    "spends": "spends",
    "Conversion Sales": "conversion_sales",
    "Conversions": "conversions",
    "signup": "signup"
})
```

```python
from sqlalchemy import create_engine

# MySQL 연결 정보 입력
user = "root"
password = "9703"
host = "localhost"  # 또는 DB서버 주소
port = 3306
database = "performance_marketing"

# SQLAlchemy 연결 엔진 생성
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}?charset=utf8mb4")

# 테이블에 DataFrame 데이터 삽입
df_performance.to_sql(
    name="performance",        # 테이블명
    con=engine,                # 연결 엔진
    if_exists="append",        # append: 기존 테이블에 추가
    index=False                # DataFrame 인덱스 컬럼 제외
)
```
